```python
import tkinter as tk
from tkinter import messagebox

MEALS = {
    'Chicken Soup': 10.0,
    'Burger': 8.0,
    'Pizza': 12.0,
    'Salad': 7.5,
}

class WaiterApp:
    def __init__(self, root):
        self.root = root
        self.total = 0.0
        self.ordered_meals = []

        label = tk.Label(root, text="Welcome to our Restaurant!")
        label.pack()

        for meal, price in MEALS.items():
            button = tk.Button(root, text=f'{meal} (${price})', command=lambda meal=meal, price=price: self.add_to_order(meal, price))
            button.pack()
        
        self.order_listbox = tk.Listbox(root)
        self.order_listbox.pack()

        remove_button = tk.Button(root, text='Remove Meal', command=self.remove_meal)
        remove_button.pack()

        total_button = tk.Button(root, text='Show Total', command=self.show_total)
        total_button.pack()

    def add_to_order(self, meal, price):
        self.total += price
        self.ordered_meals.append(meal)
        self.order_listbox.insert(tk.END, meal)

    def show_total(self):
        order_summary = "\n".join((f"{meal}: ${MEALS[meal]}"
                                   for meal in set(self.ordered_meals))) + f"\n\nTotal: ${self.total}"
        messagebox.showinfo('Total cost', order_summary)

    def remove_meal(self):
        selected_idx = self.order_listbox.curselection()
        if selected_idx:
            meal = self.order_listbox.get(selected_idx)
            self.total -= MEALS[meal]
            self.ordered_meals.remove(meal)
            self.order_listbox.delete(selected_idx)

if __name__ == "__main__":
    root = tk.Tk()
    app = WaiterApp(root)
    root.mainloop()
```
